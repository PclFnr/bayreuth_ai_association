```python
import os

import pandas as pd
```

```python
file_path_checkout = "./Kassendaten BT 2021-2024.txt"
file_path_production = "./DispoRez_Exp-2023u2024.csv"
```

```python
with open(file_path_checkout, 'r') as file:
    for i in range(0,5):  # Read the first 5 lines
        print(file.readline())
```

```python
with open(file_path_production, 'r') as file:
    for i in range(0,5):  # Read the first 5 lines
        print(file.readline())
```

```python
# Open the file and inspect the content to skip metadata lines
with open(file_path_checkout, 'r', encoding='utf-8') as file:
    lines = file.readlines()

# Determine where the table starts (e.g., line with column headers)
start_idx = next(i for i, line in enumerate(lines) if "Satz" in line)

# Extract the header and table data
header = lines[start_idx].strip()
data = lines[start_idx + 1:]

# Create a DataFrame using pandas.read_fwf for fixed-width data
df_checkout = pd.read_fwf(
    filepath_or_buffer=file_path_checkout,
    skiprows=start_idx + 1,  # Skip metadata and header rows
    header=None,             # No header in the table rows yet
    decimal=',',             # Set the decimal separator
    encoding='utf-8'         # Handle encoding issues
)

# Assign column names manually (if needed)
columns = header.split()  # Extract column names from the header line
df_checkout.columns = columns[:len(df_checkout.columns)]  # Ensure proper column assignment

# Display the resulting DataFrame
df_checkout.head()
```

```python
df_production = pd.read_csv(
    file_path_production,
    encoding='utf-8-sig',  # Handle BOM
    delimiter=';',         # Set the correct delimiter
    decimal=',',           # Set the decimal separator
    quoting=3              # Avoid issues with quotes
)
```
