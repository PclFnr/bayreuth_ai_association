```python
df_coffee = df[df['Gebinde'] == 'Kaffee']
df_coffee.head()
```

```python
df_coffee_by_day = df_coffee.groupby(pd.Grouper(key='Datum Arbeitstag', freq='D')).sum()
df_coffee_by_day.head(3)
```

```python
df_coffee_by_day.plot(y="Menge")
```
